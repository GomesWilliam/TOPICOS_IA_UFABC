## Importando Dataset

```python
dataset1 = pd.read_csv('./datasets/dataset1.txt', header = None, names=['feat1', 'feat2', 'target']) 
dataset2= pd.read_csv('./datasets/dataset2.txt', header = None, names=['feat1', 'feat2', 'target']) 
dataset3 = pd.read_csv('./datasets/dataset3.txt', header = None, names=['feat1', 'feat2', 'target']) 
dataset4 = pd.read_csv('./datasets/dataset4.txt', header = None, names=['feat1', 'feat2', 'target']) 
dataset5 = pd.read_csv('./datasets/dataset5.txt', header = None, names=['feat1', 'feat2', 'target']) 
dataset6 = pd.read_csv('./datasets/dataset6.txt', header = None, names=['feat1', 'feat2', 'target']) 
```

## Dataset1

### Plot do dataset1

```python
params = {'figure.figsize': [10, 5], 
          'axes.labelsize': 12,
          'axes.titlesize':16, 
          'font.size': 12,
          'legend.fontsize': 12, 
          'xtick.labelsize': 10, 
          'ytick.labelsize': 10
         }
plt.rcParams.update(params)

plt.scatter(dataset1["feat1"], dataset1["feat2"], c=dataset1["target"], s=150, alpha=0.80)
plt.title('Dataset1')
plt.xlabel('Caractrística1')
plt.ylabel('Caractrística2')
plt.grid(True, alpha=0.5)
plt.show()

#plt.savefig('./DataSetPlot/dataset3.pdf', format='pdf', bbox_inches = 'tight', dpi = 400)
```

### Separando amostras de treino e teste

```python
classes1 = [1 if x == 1 else 0 for x in dataset1.target]
xTrain, xTest, yTrain, yTest = tts(dataset1[['feat1', 'feat2']].values, classes1, shuffle = True, test_size = 0.25)
```

### Treinando o modelo

```python
percep1 = Perceptron(learning_rate = 0.00001, epochs = 1000)
percep1.fit(xTrain, yTrain)
percep2 = Perceptron(learning_rate = 0.0001, epochs = 1000)
percep2.fit(xTrain, yTrain)
percep3 = Perceptron(learning_rate = 0.001, epochs = 1000)
percep3.fit(xTrain, yTrain)
percep4 = Perceptron(learning_rate = 0.01, epochs = 1000)
percep4.fit(xTrain, yTrain)
percep5 = Perceptron(learning_rate = 0.1, epochs = 1000)
percep5.fit(xTrain, yTrain)
percep6 = Perceptron(learning_rate = 1, epochs = 1000)
percep6.fit(xTrain, yTrain)
```

### Evolução das Acurácias ao longo das Épocas

### Considerações

Para o dataset do 1 ao 4 foi utilizado a classe Perceptron, na qual é usada apenas para duas classes, já que sua saída é binária.
O dataset 5 e 6 foi utilizado outra classe, o MCPerceptron, na qual pode ser usado para mais de 2 classes já que ele cria um neuronio para cada classe.
A implementação de ambas foi feita por mim e para o Multiclass Perceptron foi utilizado o One X Rest, ou seja, para cada classe ele ele treina o modelo binário e após feito o treinamento para cada classe ele computa a maior saída para cada treinamento, a saída y que for maior ele ativa o neurônio da classe respectiva. 

Quanto aos datasets, vemos que datasets que são linearmente separados obtemos melhores acurácias e mesmo quanto temos mais de uma classe isso ocorre pois temos mais de uma fronteira de decisão. 
É interessante notar que o número de épocas é muito importante para se obter a convergência do algoritmo, ao longo das épocas os pesos e a acurácia vão convergindo e isso nos leva a questão de uma possibilidade para o critério de parada, após um tempo e o algoritmo não demonstrar melhora e estiver aproximadamente linear a variação dos pesos e acurácia pode ser uma solução para terminar o treinamento e economizar recursos computacionais, mas a determinação do número de épocas é dificil, para alguns problemas converge rápidamente e para outros demoram muito. 
Falando em convergência do algoritmo para o critério de parada temos um outro problema, o passo que escolhemos para a atualização dos pesos, como é observado nos gráficos acima, quanto maior é o passo do algoritmo mais variações temos nos pesos e assim mais variações nas acurácia, ficando assim dificil usar esse critério de parada, ou seja, para cada época os valores mudam muito ficando dificil poder terminar o algoritmo esperando uma boa acurácia, logo, pelo que foi visto aqui é bom escolher um número menor para o passo, porém deve-se se atenrar que precisará de mais épocas para o algoritmo convergir já que as atualizações são pequenas nos pesos, mas em compensação as variações dos pesos não poucas.
