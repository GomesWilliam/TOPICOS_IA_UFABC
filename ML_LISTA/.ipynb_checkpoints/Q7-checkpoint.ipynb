## Criando vetor de características

```python
def createDataList(data):
    dataList = []
    for i in range(len(data.feat1.values)):
        dataList.append([data.feat1.values[i],data.feat2.values[i]])
    return dataList

dataset1Data = createDataList(dataset1)
dataset2Data = createDataList(dataset2)
dataset3Data = createDataList(dataset3)
dataset4Data = createDataList(dataset4)
dataset5Data = createDataList(dataset5)
dataset6Data = createDataList(dataset6)
```

## Função Classificação

```python
def classificar(data, classData):
    
    X_train, X_test, y_train, y_test = train_test_split(data, classData, test_size=0.30)
    LDA = LinearDiscriminantAnalysis()
    QDA = QuadraticDiscriminantAnalysis()
    LDA.fit(X_train, y_train)
    QDA.fit(X_train, y_train)
    LDAPredict = LDA.predict(X_test)
    QDAPredict = QDA.predict(X_test)
    return y_test, LDAPredict, QDAPredict
```

## Classificação

```python
y_test1, LDAPredict1, QDAPredict1 = classificar(dataset1Data, dataset1.target)
y_test2, LDAPredict2, QDAPredict2 = classificar(dataset2Data, dataset2.target)
y_test3, LDAPredict3, QDAPredict3 = classificar(dataset3Data, dataset3.target)
y_test4, LDAPredict4, QDAPredict4 = classificar(dataset4Data, dataset4.target)
y_test5, LDAPredict5, QDAPredict5 = classificar(dataset5Data, dataset5.target)
y_test6, LDAPredict6, QDAPredict6 = classificar(dataset6Data, dataset6.target)
```

## Métricas

```python
#DATASET1 ACURÁCIA PARA O LDA E QDA RESPECTIVAMENTE
AccLDA1 = accuracy_score(y_test1, LDAPredict1)
AccQDA1 = accuracy_score(y_test1, QDAPredict1)

#DATASET2 ACURÁCIA PARA O LDA E QDA RESPECTIVAMENTE
AccLDA2 = accuracy_score(y_test2, LDAPredict2)
AccQDA2 = accuracy_score(y_test2, QDAPredict2)

#DATASET3 ACURÁCIA PARA O LDA E QDA RESPECTIVAMENTE
AccLDA3 = accuracy_score(y_test3, LDAPredict3)
AccQDA3 = accuracy_score(y_test3, QDAPredict3)

#DATASET4 ACURÁCIA PARA O LDA E QDA RESPECTIVAMENTE
AccLDA4 = accuracy_score(y_test4, LDAPredict4)
AccQDA4 = accuracy_score(y_test4, QDAPredict4)

#DATASET5 ACURÁCIA PARA O LDA E QDA RESPECTIVAMENTE
AccLDA5 = accuracy_score(y_test5, LDAPredict5)
AccQDA5 = accuracy_score(y_test5, QDAPredict5)

#DATASET6 ACURÁCIA PARA O LDA E QDA RESPECTIVAMENTE
AccLDA6 = accuracy_score(y_test6, LDAPredict6)
AccQDA6 = accuracy_score(y_test6, QDAPredict6)

print("Métricas datase1.txt")
print("  Acurácia LDA "+ str(round(AccLDA1*100)) + "%")
print("  Acurácia QDA "+ str(AccQDA1*100) + "%")
print("-----------------------")
print("Métricas datase2.txt")
print("  Acurácia LDA "+ str(round(AccLDA2*100, 2)) + "%")
print("  Acurácia QDA "+ str(round(AccQDA2*100, 2)) + "%")
print("-----------------------")
print("Métricas datase3.txt")
print("  Acurácia LDA "+ str(round(AccLDA3*100, 2)) + "%")
print("  Acurácia QDA "+ str(round(AccQDA3*100, 2)) + "%")
print("-----------------------")
print("Métricas datase4.txt")
print("  Acurácia LDA "+ str(round(AccLDA4*100, 2)) + "%")
print("  Acurácia QDA "+ str(round(AccQDA4*100, 2)) + "%")
print("-----------------------")
print("Métricas datase5.txt")
print("  Acurácia LDA "+ str(round(AccLDA5*100, 2)) + "%")
print("  Acurácia QDA "+ str(round(AccQDA5*100, 2)) + "%")
print("-----------------------")
print("Métricas datase6.txt")
print("  Acurácia LDA "+ str(round(AccLDA6*100, 2)) + "%")
print("  Acurácia QDA "+ str(round(AccQDA6*100, 2)) + "%")
```
