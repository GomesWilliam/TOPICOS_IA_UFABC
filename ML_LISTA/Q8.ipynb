## Importando Bibliotecas usada no problema

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

## Importando Dataset

```python
dataset = pd.read_csv('./datasets/dataset2.txt', header = None, names=['feat1', 'feat2', 'target']) 
```

## Plotando Dataset

```python
b = sns.scatterplot(x="feat1", y="feat2", hue='target', data=dataset, markers=True, style="target",  palette="Set1", legend = 'full')
b.set_title('Dataset 2')
b.set_xlabel('Característica 1')
b.set_ylabel('Característica 2') 
```

## Criando vetor de características

```python
def createDataList(data):
    dataList = []
    for i in range(len(data.feat1.values)):
        dataList.append([data.feat1.values[i],data.feat2.values[i]])
    return dataList

datasetData = createDataList(dataset)
```

## Classificação

```python
coefs = []
intercept = []
for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(datasetData, dataset.target, test_size=20)
    LDA = LinearDiscriminantAnalysis()
    LDA.fit(X_train, y_train)
    LDAPredict = LDA.predict(X_test)
    AccLDA = accuracy_score(y_test, LDAPredict)
    coefs.append(LDA.coef_[0])
    intercept.append(LDA.intercept_)
    print("Subamostragem "+ str(i+1))
    print("Acurácia LDA "+ str(round(AccLDA*100)) + "%")
    print("-----------------------")
```

```python
## Fronteira de decisões
#equaçãos da reta

x1= np.arange(-30,30,0.5)

x2_1 = (coefs[0][0]*x1 + intercept[0])/coefs[0][1]
x2_2 = (coefs[1][0]*x1 + intercept[1])/coefs[1][1]
x2_3 = (coefs[2][0]*x1 + intercept[2])/coefs[2][1]
x2_4 = (coefs[3][0]*x1 + intercept[3])/coefs[3][1]
x2_5 = (coefs[4][0]*x1 + intercept[4])/coefs[4][1]
x2_6 = (coefs[5][0]*x1 + intercept[5])/coefs[5][1]
x2_7 = (coefs[6][0]*x1 + intercept[6])/coefs[6][1]
x2_8 = (coefs[7][0]*x1 + intercept[7])/coefs[7][1]
x2_9 = (coefs[8][0]*x1 + intercept[8])/coefs[8][1]
x2_10 = (coefs[9][0]*x1 + intercept[9])/coefs[9][1]



plt.figure(figsize=(7,7))

plt.plot(x1,x2_1, color = 'red', label = r'subamostragem 1')
plt.plot(x1,x2_2, color='green', linestyle='dashed', label = r'subamostragem 2')
plt.plot(x1,x2_3, color='blue', linestyle='dashed', label = r'subamostragem 3')
plt.plot(x1,x2_4, color='cyan', linestyle='dashed', label = r'subamostragem 4')
plt.plot(x1,x2_5, color='magenta', linestyle='dashed', label = r'subamostragem 5')
plt.plot(x1,x2_6, color='yellow', linestyle='dashed', label = r'subamostragem 6')
plt.plot(x1,x2_7, color='black', linestyle='dashed', label = r'subamostragem 7')
plt.plot(x1,x2_8, color=(0,1,0,1), linestyle='dashed', label = r'subamostragem 8')
plt.plot(x1,x2_9, color=(0.85,1,0.3,0.7), label = r'subamostragem 9')
plt.plot(x1,x2_10, color='red', linestyle='dashed', label = r'subamostragem 10')
sns.scatterplot(x="feat1", y="feat2", hue='target', data=dataset, markers=True, style="target",  palette="Set1", legend = False)






plt.xlabel(r'$x_{1}$', fontsize = 15)
plt.ylabel(r'$x_{2}$', fontsize = 15)

plt.xlim(-3,3)
plt.ylim(-2,2)

#plt.xlim(-2,2)
#plt.ylim(-2,2)
plt.legend()
#plt.axis('equal')
plt.show()
```
